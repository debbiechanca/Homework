```python
# Lists for holding sentiments
compound_list = []
date_list = []

news_org_df = pd.DataFrame()

# Target news organization
news_orgs = ("@BBC", "CBS")
```

```python
for target_org in news_orgs:

    # Retrieve 100 most recent tweets
    public_tweets = api.search(
        target_org,
        count=100,
        result_type="recent",
        lang="en")
    
    # Print Tweets
    for tweet in public_tweets["statuses"]:
        # Parse the tweet to identify its text and print
        sentiment_string = tweet["text"]
        tweet_id = tweet["id_str"]
        
        created_date = tweet["created_at"]
                
        print ("Created Date: " , created_date)
        print (sentiment_string.encode("UTF-8"))

        # Analyze the sentiment of the tweet
        compound = analyzer.polarity_scores(sentiment_string)["compound"]

        # Add the sentiment analyses for compound polarity to the list
        compound_list.append(compound)
        date_list.append(created_date)
        
```
