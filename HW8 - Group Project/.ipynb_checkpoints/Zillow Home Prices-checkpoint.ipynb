```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Open csv files for reading City Survey Master Data 1996-2017 csv file downloaded from http://sfgov.org/citysurvey/file/95.
# Save path to data set in a variable
data_file = "Resources/Zip_Zhvi_SingleFamilyResidence.csv"
```

```python
# Use Pandas to read data
home_df = pd.read_csv(data_file)
home_df.head()
```

```python
# Select San Francisco data
# Select only 2017 home prices
# Get aveage home price for 2017

sfhome_df = home_df[home_df.City == 'San Francisco']
sfhome_df.head()
```

```python
sfhome_df.dtypes
```

```python
sfhome_df.rename(
  columns={
    '2017-01' : 'Jan2017',
    '2017-02' : 'Feb2017',
    '2017-03' : 'Mar2017',
    '2017-04' : 'Apr2017',
    '2017-05' : 'May2017',
    '2017-06' : 'Jun2017',
    '2017-07' : 'Jul2017',
    '2017-08' : 'Aug2017',
    '2017-09' : 'Sep2017',
    '2017-10' : 'Oct2017',
    '2017-11' : 'Nov2017',
    '2017-12' : 'Dec2017'
  },
  inplace=True
)

AvePrice = sfhome_df[['Jan2017', 'Feb2017', 'Mar2017', 'Apr2017', 'May2017', 'Jun2017', 'Jul2017', 'Aug2017', 'Sep2017', 'Oct2017', 'Nov2017', 'Dec2017']].mean(axis=1)

sfhome_df['AvePrice'] = AvePrice
```
